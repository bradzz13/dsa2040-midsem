```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the data (Real World Dataset)

```python
#Extract raw data 
df=pd.read_csv('D:\\new_retail_data.csv\\new_retail_data.csv')
# Check unique years in the dataset
print(df['Year'].unique())

df['Year'] = df['Year'].fillna(2023)  # Assuming missing years are 2023
```

Split the dataset on date to make the raw data and incremental data for more recent data

```python
print(df['Year'].value_counts(dropna=False))

raw_df = df[df['Year'] == 2023]
incremental_df = df[df['Year'] == 2024]

print(f"Raw rows: {len(raw_df)}, Incremental rows: {len(incremental_df)}")
```

```python
# save the datasets 
raw_df.to_csv('data/raw_data.csv', index=False)
incremental_df.to_csv('data/incremental_data.csv', index=False)
```
